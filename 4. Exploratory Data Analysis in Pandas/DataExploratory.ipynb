```python
# Exclude non-numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64'])
```

```python
# numeric_columns = df.select_dtypes(include=['float64', 'int64'])
columns_to_assign = ['Continent']
temp_df = df[columns_to_assign].copy()
temp_df
# df2.groupby('Continent').mean()
```

```python
df2 = pd.concat([temp_df, numeric_columns], axis=1)
df2
```

```python
df3 = df2.groupby('Continent').mean().sort_values(by='2022 Population', ascending=False)
df3
```

```python
sns.heatmap(df3)
```
